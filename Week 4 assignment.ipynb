## Model building - Using Logistic Regression

This dataset is an example of binary classification hence we will be using logistic regression to build a model

### 1. Import model building libraries

```python
from sklearn.linear_model import LogisticRegression
```

### 2. Initialize model

```python
lr_model = LogisticRegression()
lr_model
```

### 3. Train the model

```python
lr_model.fit(x_train, y_train)
```

### 4. Test the model

```python
lr_pred = lr_model.predict(x_test)
lr_pred
```

Now we have to compare these values with y_test values

```python
y_test
```

### 5. Random value prediction

Predicted - Yes;
Actual - Yes

Hence our model seems to be working fine. But we have to run some additional codes to see how efficient our model is

### 6. Evaluation of classification model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
# accuracy score

accuracy_score(y_test, lr_pred)
```

```python
# confusion matrix 

lr_cm = confusion_matrix(y_test, lr_pred)
lr_cm
```

```python
pd.crosstab(y_test, lr_pred)
```

```python
tp = lr_cm[1][1]

tn = lr_cm[0][0]

fp = lr_cm[0][1]

fn = lr_cm[1][0]
```

```python
accuracy_score = (tp+tn)/(tp+tn+fp+fn)
accuracy_score
```

```python
precision = tp/(tp+fp)
precision
```

```python
recall = tp/(tp+fn)
recall
```

```python
f1_score = 2*(precision*recall)/(precision+recall)
f1_score
```

### 7. ROC AUC curve

```python
lr_prob = lr_model.predict_proba(x_test)[:, 1]
lr_prob
```

```python
fpr, tpr, threshold = roc_curve(y_test, lr_prob, pos_label='Yes')
```

```python
fpr
```

```python
tpr
```

```python
threshold
```

```python
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression')
plt.show()
```

For logistic regression -

The model can be improved significantly

We shall use hyperparameter tuning to do so

```python
from sklearn.model_selection import GridSearchCV
```

```python
# can get this from the documentation
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  
    'penalty': ['l1', 'l2'],              
    'solver': ['liblinear', 'saga']       
}
```

```python
lr_grid_search = GridSearchCV(estimator = lr_model, param_grid = param_grid, cv = 5, scoring='f1', verbose = 1, n_jobs = -1)
lr_grid_search
```

```python
lr_grid_search.fit(x_train, y_train)
```

```python
lr_grid_search.best_params_
```

```python
lr_cv = LogisticRegression(
    C=0.001,
    penalty='l1',  
    solver='liblinear'
)
lr_cv
```

```python
lr_best_model = lr_grid_search.best_estimator_
lr_best_model
```

```python
lr_pred = lr_best_model.predict(x_test)
lr_pred
```

```python
y_test
```

```python
print(classification_report(y_test, lr_pred))
```

As we can see, after hyperparameter tuning, accuracy has reduced from 83% to 80% so hyperparameter tuning is not required or we can use different parameters and test it again.
