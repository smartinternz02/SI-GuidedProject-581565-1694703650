The dataset gives us statistics based on the percentage of car crashes, what caused them and insurance details in a particular states in the US

## Graph 1 - To compare the premium insurance values for each state

```python
sns.barplot(x="abbrev", y="ins_premium", data=data, width=1)
plt.xticks(rotation=90)
plt.title("Insurance Premium by State")
```

Inference - 

Premium insurance is moderately priced in most states with a few exceptions 
1. State with most premium insurances - New Jersey (NJ) followed by Los Angeles (LA) and Washington DC (DC)
2. State with least premium insurance - Hawaii (HI) followed by Iowa (IA) and Maine (ME

## Graph 2 - To compare the insurance losses for each state

```python
sns.barplot(x="abbrev", y="ins_losses", data=data, width=1)
plt.xticks(rotation=90)
plt.title("Insurance Losses by State")
```

Inference - 

Insurance losses are definitely on the higher side
1. Los Angeles (LA), Maryland (MD) and Oklahoma (OK) have suffered highest insurance losses i.e, between $175k-200k
   
2. Idaho (ID) and Missouri (MO) have the least insurance losses i.e, between $75k-100k

## Graph 3 - To visualize trend in crashes due to speeding in different states

```python
sns.lineplot(x="abbrev", y="speeding", data=data)
plt.xticks(rotation=90)
plt.title("Crashes due to speeding by State")
```

Inference -

The percentage of crashes due to speeding are slightly on the higher side

## Graph 4 - To visualize trend in crashes due to alcohol in different states

## Graph 5 - To study the relationship between car crashes due to alcohol and insurance losses

Inference - 

There is no definite pattern observed when it comes to insurance losses due to alcohol driven car crashes
1. 0-2% car crashes incur insurance losses upto $140k

2. 2-4% car crashes incur insurance losses upto $195k (roughly)

3. 4-6% car crashes incur insurance losses upto $180k
   
4. 6-8% car crashes incur comparatively lesser insurance losses. However it does go upto $198k (roughly)
5. 8-10% car crashes incur the least insurance losses (comparatively)

## Graph 6 - To study the distribution of premium insurances in all states

```python
sns.histplot(data["ins_premium"], bins=10, kde=True)
plt.xticks(rotation=90)
plt.title("Insurance Premium Distribution")
```

Inferences - 

1. Central Tendency: It provides an idea of where the majority of insurance premiums fall (peak of histogram). It appears that the mode is somewhere around $800k.

2. Skewness: This is a postively skewed distribution

3. Outliers: Outliers are values that significantly deviate from the majority of data points. In this histogram, any data points that fall far to the right of the main peak could be considered outliers i.e, 900-1000 and 1200.

4. Bimodality: A bimodal distribution has two distinct peaks, indicating that there might be two subgroups within the data. In this case, the histogram appears unimodal, meaning it has one main peak and only one subgroup.

5. Spread: Wider distributions indicate higher variability, while narrower distributions imply lower variability. In this case, the distribution appears somewhat wide, suggesting moderate variability in insurance premiums.

6. Kurtosis: Kurtosis measures the "tailedness" of the distribution. A higher kurtosis indicates heavy tails, while lower kurtosis indicates light tails. In this histogram, it appears to have moderate kurtosis.

## Graph 7 - To visualize the percentage of insurance losses

```python
sizes = data["ins_losses"]
labels = data["abbrev"]
fig = plt.figure(figsize=(14, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Percentage of Insurance losses")
```

Inference - 

Insurance losses are mostly uniform in all states

1. Highest losses - Los Angeles (LA), Oklahoma (OK), Connecticut (CT)
2. Least losses - Montana (MT) and Idaho (ID)

## Graph 8 - To visualize the correlations between the numeric values

```python
new_data = data.iloc[:, :-1]
correlation_matrix = new_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap")
```

The color intensity of each cell represents the strength and direction of the correlation - 
1. Dark blue indicates a strong negative correlation
2. Dark red indicates a strong positive correlation
3. Light colors indicate weaker or no correlation

Inferences -

1. Speeding vs. Alcocol: Light colour indicates there is no clear relationship between the two variables

2. Speeding vs. Not Distracted: Light colour indicates there is no clear relationship between the two variables
  
3. Alcohol vs. Not Distracted: Light colour indicates there is no clear relationship between the two variables

4. Ins Premium vs. Ins Losses: Light colour indicates there is no clear relationship between the two variables
   
5. Total vs. Ins Premium: Dark blue indicates states with higher crash rates tend to have lesser premium insurance and vice versalosses.
